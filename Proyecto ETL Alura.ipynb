## ðŸ“Œ ExtracciÃ³n

```python
import pandas as pd
url = "https://raw.githubusercontent.com/alura-cursos/challenge2-data-science-LATAM/main/TelecomX_Data.json"
data = pd.read_json(url)
```

```python
data.head(1)
```

```python
df = pd.json_normalize(data.to_dict(orient="records"))
df.columns = df.columns.str.replace('.','_')
df.drop('customerID',axis=1,inplace=True)
def quitar_prefijo(lista):
    return [col.split('_', 1)[-1] for col in lista]
df.columns = quitar_prefijo(df.columns)
df.head(3)
```

```python
df = df[df['Churn']!='']
```

```python
for col in df.columns:
       if df[col].nunique()>2:
        if df[col].nunique()<10:

          print(f'valores de la columna {col}: {df[col].nunique()}')
          print(df[col].unique())
```

## ðŸ”§ TransformaciÃ³n

```python
for col in ['PaperlessBilling','Partner','Dependents','PhoneService','gender','Churn']:
  df[col] = df[col].replace({'Yes':1,'No':0, 'Female':0,'Male':1}).astype(int)
```
